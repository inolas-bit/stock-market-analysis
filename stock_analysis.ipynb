```python
# Import libraries
import yfinance as yf
import plotly.graph_objects as go
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# for interactive inline 
%matplotlib inline 
```

```python
#download data (NSE  stocks )
tickers = ['RELIANCE.NS','TCS.NS','INFY.NS']
data = {}

for ticker in tickers:
    data[ticker]=yf.download(ticker,start="2023-01-01",end="2024-12-31")
```

```python
#create dataframes for each 
reliance = data['RELIANCE.NS'].reset_index()
tcs = data['TCS.NS'].reset_index()
infy = data['INFY.NS'].reset_index()
```

```python

#Calculate moving avg
reliance['MA_20'] = reliance['Close'].rolling(window=20).mean()
reliance['MA_50'] = reliance['Close'].rolling(window=50).mean()


tcs['MA_20'] = tcs['Close'].rolling(window=20).mean()
tcs['MA_50'] = tcs['Close'].rolling(window=50).mean()



infy['MA_20'] = infy['Close'].rolling(window=20).mean()
infy['MA_50'] = infy['Close'].rolling(window=50).mean()
```

```python
# calculate daily return & volatility
reliance['Daily Return'] = reliance['Close'].pct_change()
reliance['Volatility_20D'] = reliance['Daily Return'].pct_change()


tcs['Daily Return'] = tcs['Close'].pct_change()
tcs['Volatility_20D'] = tcs['Daily Return'].pct_change()


infy['Daily Return'] = infy['Close'].pct_change()
infy['Volatility_20D'] = infy['Daily Return'].pct_change()
```

```python
import matplotlib.pyplot as plt

# Assuming already calculated 'MA_20' and 'MA_50' for the 'reliance' DataFrame
#visualizing price + moving volatility 

plt.figure(figsize=(12, 6))
plt.plot(reliance['Date'], reliance['Close'], label='Close Price')
plt.plot(reliance['Date'], reliance['MA_20'], label='20-Day MA')
plt.plot(reliance['Date'], reliance['MA_50'], label='50-Day MA')
plt.title("Reliance Stock Price with Moving Averages")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show() 
```

```python
import matplotlib.pyplot as plt

# Assuming already calculated 'MA_20' and 'MA_50' for the 'tcs' DataFrame
#visualizing price + moving volatility 

plt.figure(figsize=(12, 6))
plt.plot(tcs['Date'], tcs['Close'], label='Close Price')
plt.plot(tcs['Date'], tcs['MA_20'], label='20-Day MA')
plt.plot(tcs['Date'], tcs['MA_50'], label='50-Day MA')
plt.title("TCS Stock Price with Moving Averages")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show() 
```

```python
#plot daily return and volatility 

fig, ax = plt.subplots(2, 1, figsize=(12, 8), sharex=True)
ax[0].plot(reliance['Date'], reliance['Daily Return'], color='grey')
ax[0].set_title('Reliance Daily Return')
ax[1].plot(reliance['Date'], reliance['Volatility_20D'], color='red')
ax[1].set_title('Reliance 20-Day Rolling Volatility')
plt.xlabel("Date")
plt.tight_layout()
plt.show()
```

```python

import pandas as pd

# Create a DataFrame of closing prices
reliance_close = reliance.set_index('Date')['Close']
tcs_close = tcs.set_index('Date')['Close']
infy_close = infy.set_index('Date')['Close']

# Concatenate the Series, aligning indices
closing_prices = pd.concat(
    [reliance_close, tcs_close, infy_close],
    axis=1,
    keys=['Reliance', 'TCS', 'INFY'],
    join='inner'  # Important: Use 'inner' join to align on common dates
)

# Calculate the correlation matrix of percentage changes
correlation = closing_prices.pct_change().corr()

print("Correlation Matrix:")
print(correlation)


    
    
```
